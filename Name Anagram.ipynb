```python
df = pd.read_csv(path+'\yob'+str(1880)+'.txt',names=['name','gender','count'])
df['year'] = date

for date in range(1881,2018):
    df_temp = pd.read_csv(path+'\yob'+str(date)+'.txt',names=['name','gender','count'])
    df_temp['year'] = date
    df = df.append(df_temp)
df_names = pd.DataFrame(df.groupby(['name','gender','year'])['count'].sum())    
```

```python
df_names.reset_index(inplace=True)
df_names.columns = ['name', 'gender','year','count']
```

```python
plt.plot(x=, df_names[(df_names['name']=='Jordan')&(df_names['gender']=='F')]['count'].values)
```

```python
name = 'Gregor'
plt.plot(df_names[(df_names['name']==name)&(df_names['gender']=='F')]['year'].values,df_names[(df_names['name']==name)&(df_names['gender']=='F')]['count'].values,label='Female')
plt.plot(df_names[(df_names['name']==name)&(df_names['gender']=='M')]['year'].values,df_names[(df_names['name']==name)&(df_names['gender']=='M')]['count'].values,label='Male')
plt.legend()
plt.show()
```
